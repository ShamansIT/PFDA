```python
# Import lib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from CSV-file
df = pd.read_csv('weatherreadings1.csv')
```

```python
df.info()
```

```python
# Convert 'reportStartDateTime' column to datetime
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])
```

```python
# Select the relevant columns: reportStartDateTime and dryBulbTemperature_Celsius
temperature_df = df[['reportStartDateTime', 'dryBulbTemperature_Celsius']]
```

```python
# Plot temperature over time
plt.figure(figsize=(10,6))
plt.plot(temperature_df['reportStartDateTime'], temperature_df['dryBulbTemperature_Celsius'], marker='o', linestyle='-', color='blue')
plt.title('Dry Bulb Temperature (Celsius) Over Time', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
